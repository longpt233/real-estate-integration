```python
import pandas as pd
```

```python
PATH_ALO = "/content/drive/MyDrive/data-standard/alonhadat.csv"
alo = pd.read_csv(PATH_ALO, encoding = 'utf-8') 
PATH_BAT_DONG_SAN_123 = "/content/drive/MyDrive/data-standard/batdongsan123.csv"
batdongsan123 = pd.read_csv(PATH_BAT_DONG_SAN_123, encoding = 'utf-8')
PATH_I = "/content/drive/MyDrive/data-standard/ibatdongsan.csv"
ibat = pd.read_csv(PATH_I, encoding = 'utf-8')
PATH_24 = "/content/drive/MyDrive/data-standard/nhadat24h.csv"
nha24 = pd.read_csv(PATH_24, encoding = 'utf-8')
```

```python
alo_min=alo[["bedroom","code","date","diningroom","floor","kitchen","length","phone_contact","price","road_width","square","width"]]
bat_min=batdongsan123[["acreage","bathroom","bedroom","code","date","floor","phone_contact","price"]]
ibat_min=ibat[["area","bedroom","code","date","diningroom","floor","length","phone_contact","price","width","world_highway"]]
nha24_min=nha24[["bathroom","bedroom","code","date","floor","ground_area","kitchen","length","livingroom","price","road_width","usable_area","width"]]
```

Chọn các trường có kiểu dữ liệu dạng số bằng cách liệt kê thủ công

```python
alo_col = list(alo_min.columns)
batdongsan123_col = list(bat_min.columns)
ibat_col = list(ibat_min.columns)
nha24_col = list(nha24_min.columns)
alo_col.sort()
batdongsan123_col.sort()
ibat_col.sort()
nha24_col.sort()
```

Tạo một List 1 chiều chứa tên trường của các bảng dữ liệu, sau đó sắp xếp theo thứ tự bảng chữ cái

```python
print(alo_col)
print(batdongsan123_col)
print(ibat_col)
print(nha24_col)
```

```python
from strsimpy.normalized_levenshtein import NormalizedLevenshtein
levenshtein=NormalizedLevenshtein()
print("Matching alonhadat vs batdongsan123")
df=pd.DataFrame(columns=['alonhadat','batdongsan123','distance'])
for i in range(0,len(alo_col)):
  for j in range(0,len(batdongsan123_col)):
    new_row={'alonhadat':alo_col[i], 'batdongsan123':batdongsan123_col[j], 'distance': levenshtein.distance(alo_col[i],batdongsan123_col[j])}
    df=df.append(new_row, ignore_index=True)
df2=df.sort_values(by=['distance'], ascending=True)
df2.head(12)
```

Matching từng tên trường của alonhadat với tất cả tên trường của batdongsanso sử dụng hàm khoảng cách Normalized Levenshtein, output là một dataframe

```python
from strsimpy.normalized_levenshtein import NormalizedLevenshtein
levenshtein=NormalizedLevenshtein()
print("Matching alonhadat vs ibatdongsan")
df=pd.DataFrame(columns=['alonhadat','ibatdongsan','distance'])
for i in range(0,len(alo_col)):
  for j in range(0,len(ibat_col)):
    new_row={'alonhadat':alo_col[i], 'ibatdongsan':ibat_col[j], 'distance': levenshtein.distance(alo_col[i],ibat_col[j])}
    df=df.append(new_row, ignore_index=True)
df2=df.sort_values(by=['distance'], ascending=True)
df2.head(15)
```
